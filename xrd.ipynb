Déterminer les 3 premiers pics du diffractogramme correspondant (en indiquant les indices hkl associés) pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å.

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import *
from matplotlib import pyplot
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
with MPRester("2HJyrqCBxbcTKw7d") as m:
    structure = m.get_structure_by_material_id("mp-23209")    
    
calc = XRDCalculator(wavelength = 1.54060 )
pattern = calc.get_pattern(structure, scaled=True)
```

```python
print("1er pic:")
print("Angle 2Θ:", pattern.x[0])
print("Intensité:", pattern.y[0])
print("hkl:", pattern.hkls[0])
print("d:", pattern.d_hkls[0], '\n')

print("2ieme pic:")
print("Angle 2Θ:", pattern.x[1])
print("Intensité:", pattern.y[1])
print("hkl:", pattern.hkls[1])
print("d:", pattern.d_hkls[1], '\n')

print("3ieme pic:")
print("Angle 2Θ:", pattern.x[2])
print("Intensité:", pattern.y[2])
print("hkl:", pattern.hkls[2])
print("d:", pattern.d_hkls[2])
```

```python
calc.show_plot(structure, two_theta_range=(0,40))
```
